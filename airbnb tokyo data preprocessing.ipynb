### Amenities

```python
ets = tokyo.loc[:,"amenities"]
```

```python
lets = []
for r in ets:
    lets.append(r.lower())
```

```python
sw = set(["[", ", ", "]"])
```

```python
king = []
```

```python
for i in range(0,len(lets)):
    queen = []
    for w in lets[i].split('"'):
        if w not in sw:
            if '\\' not in list(w):
                queen.append(w)
    king.append(queen)
```

코드 해석

king이라는 array는 사전에 만들어진 상황.
일단 첫번째 반복은 lets 속 요소 개수만큼. 그게 인덱스 수랑 똑같지. 숙소 개수랑.
숙소 하나당 [] 하나 배정해서 그 안에 amenity를 담는 게 목표지.

두 번째 for 문에서는 각 숙소 내에서 일어남. lets[i].split('"') 이렇게 했을 때 단어와 [,],,가 혼재되어 있으니 기호들을 제거하는 과정. 기호들을 미리 불용어로 지정해놓고, 불용어인 애들은 빼는 게 if w not in sw:.

그밑은 \\가 포함된 단어를 제거하는 과정. list(w)하면 철자별로 쪼개지기 때문에, \\가 포함된 단어는 queen에 포함을 안 시켜버림.

## 잠깐 amenities 탐색

```python
queen=[]
for i in range(0,len(lets)):
    for w in lets[i].split('"'):
        if w not in sw:
            if '\\' not in list(w):
                queen.append(w)
```

```python
queen
```

```python
u_words = set(queen)
```

```python
u_words
```

```python
from collections import Counter
```

```python
vector = []

for r in king:
    counter = Counter(r)
    count = [counter.get(w, 0) for w in u_words]
    vector.append(count)
```

```python
count_vector = pd.DataFrame(vector, columns=u_words)
count_vector
```

```python
count_vector.iloc[:,0].sum()
```
